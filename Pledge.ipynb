```python
import pandas as pd
import re
```

```python
df = pd.read_excel('Urls.xlsx')
df
```

```python
df[df['index'] == 499]
```

```python
df_new = df['Pledge_Text'][:500]
df_new[0]
```

```python
pledge = []
for val in range(df_new.shape[0]):
    text = df_new[val].replace("_x000D_\n",". ")
    text = text.replace(",","")
    pledge.append(text)
```

```python

pledge[2]
```

```python

min_amt = []
max_amt = []
mean = []
pledge_count = []
count = 0

check = ['Pledge','or more']

for val in pledge:
    try:
        temp_set = []
        amt_set = []
        val = '"' + val + '"'
        desc = val.split('.')
        for line in desc:
            if all([x in line for x in check]):
                temp_set.append(line)
        for str in temp_set:
            amt = min(re.findall(r'\d+', str))
            amt_set.append(amt)
        min_amt.append(min(amt_set))
        max_amt.append(max(amt_set))
        # mean.append(sum(amt_set)/len(amt_set))
        pledge_count.append(len(temp_set))
        count += 1
    except ValueError:
        print(count)
        count += 1
    
```
