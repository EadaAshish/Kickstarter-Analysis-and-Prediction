```python
urls = []
for pos in range(df.shape[0]):
    val = eval(df['category'][pos])
    urls.append(val['slug'])
```

```python
df_temp = pd.DataFrame(urls)
df_temp.describe()
```

```python
df2 = df_temp.set_axis(['Cat'], axis=1, inplace = False)
df2
```

```python
new = df2["Cat"].str.split("/", n = 1, expand = True)
```

```python
df2['Parent'] = new[0]
```
